```python
import pandas as pd

df = pd.read_csv("50_Startups.csv")

df
```

```python
df.columns
```

```python
df = pd.get_dummies(df, drop_first= True)
```

```python
df
```

```python
indp = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida','State_New York']]
depe =df[['Profit']]
```

```python
## split traing and data set

from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(indp, depe,test_size=0.33,random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
model = regressor.fit(x_train,y_train) ## y = b0+b1..._bn
```

```python
model
```

```python
bias = model.intercept_
weight = model.coef_
```

```python
print(bias)
print(weight)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score

r_score = r2_score(y_test,y_pred)
```

```python
r_score
```
